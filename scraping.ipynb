```python
import requests
import bs4, lxml
import re 
import concurrent.futures
import hashlib
import os
import time
import random
import gzip

try:
  os.mkdir('htmls')
except:
  ...

def html(number_of_pages):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.13; rv:58.0) Gecko/20100101 Firefox/58.0'}
    headers['referer'] = 'https://www.amazon.co.jp'
    url = f"https://ramendb.supleks.jp/rank?page={number_of_pages}"
    try:
          r = requests.get(url, headers=headers)
    except Exception as e:
        print(e)
        return []
    r.encoding = r.apparent_encoding
    html = r.text
    soup = bs4.BeautifulSoup(html, 'lxml')
    a_list = soup.find(class_="ranks").find_all("a", href=True)
    return ["https://ramendb.supleks.jp" + x["href"] for x in a_list if re.search(r'\/s/[0-9]+\.html', x["href"]) is not None]

def detail_html(url):
    try:
        save_name = 'htmls/' + hashlib.sha256(bytes(url,'utf8')).hexdigest()
        if os.path.exists(save_name) is True:
            return []
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.13; rv:58.0) Gecko/20100101 Firefox/58.0'}
        headers['referer'] = 'https://www.amazon.co.jp'
        try:
            r = requests.get(url, headers=headers)
            #print(r.status_code)
        except Exception as e:
            return []
        time.sleep(random.randint(2,5))
        r.encoding = r.apparent_encoding
        html = r.text
        try:
            open(save_name, 'wb').write( gzip.compress(bytes(html,'utf8')) )
        except OSError:
            return []
        return html
    except Exception as ex:
        print(ex)
        return []    

def get_max_pages():
    left,right = 1,100000
    while left < right - 1:
        middle = (left + right) // 2
        if len(html(middle)) == 0:
            right = middle
        else:
            left = middle
    print(left, right)
    
def main():
    max_pages = 657
    current_page = 1
    max_workers = 64
    
    while current_page < max_pages:
        detail_urls = set()
        with concurrent.futures.ProcessPoolExecutor(max_workers=max_workers) as executor:
            for urls in executor.map(html, range(current_page, current_page + 10)):
                for url in urls:
                    detail_urls.add(url)
        with concurrent.futures.ProcessPoolExecutor(max_workers=max_workers) as executor:
            executor.map(detail_html, detail_urls)
        print(current_page)
        current_page += 10
main()
```
