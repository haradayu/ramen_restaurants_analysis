```python
from pathlib import Path
import bs4
import gzip
import concurrent.futures
import json
import lxml
import hashlib
import os

try:
  os.mkdir('jsons')
except:
  ...

def _map(arg):
    key, names = arg
    for name in names:
        try:
            html = gzip.decompress(name.open("rb").read()).decode()
            soup = bs4.BeautifulSoup(html,"lxml")
            ramen_json = soup.find("script", {"type":"application/ld+json"})
            ramen_json = json.loads(ramen_json.get_text())
            ramen_name = ramen_json["name"]
            gegion = ramen_json["address"]["addressRegion"]
            locality = ramen_json["address"]["addressLocality"]
            street_address = ramen_json["address"]["streetAddress"]
            rating_value = ramen_json["aggregateRating"]["ratingValue"]
            review_count = ramen_json["aggregateRating"]["reviewCount"]
            ration_count = ramen_json["aggregateRating"]["ratingCount"]
            latitude = ramen_json["geo"]["latitude"]
            longitude = ramen_json["geo"]["longitude"]
            obj = {'name':ramen_name, 'gegion': gegion, 'locality':locality, 'street_address':street_address,
                    'street_address':street_address, 'rating_value':rating_value, 'review_count':review_count, 'ration_count': ration_count,
                    'latitude':latitude, 'longitude': longitude}
            json.dump(obj, fp=open(f'jsons/{name.name}.json', 'w'), indent=2, ensure_ascii=False)
            
        except Exception as ex:
            print(ex)
            ...
        

def main():
    args = {}
    for index, name in enumerate(Path('./htmls').glob('*')):
        key = index % 16
        if args.get(key) is None:
            args[key] = []
        args[key].append(name)
    args = [(key, names) for key, names in args.items()]
#     _map(args[0])
    with concurrent.futures.ProcessPoolExecutor(max_workers=32) as exe:
        exe.map(_map, args)
main()
```
